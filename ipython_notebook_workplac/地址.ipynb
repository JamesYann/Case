```python
#coding=utf-8
from urllib.request import urlopen    #導入urlopen模塊和BeautifulSoup模塊
import requests
from bs4 import BeautifulSoup

for x in range(1,4128):
    link = "https://okgo.tw/butyview.html?id="+str(x)
    r = requests.get(link) #將網頁資料GET下來
    soup = BeautifulSoup(r.text,"html.parser") #將網頁資料以html.parser
    sel = soup.select("div.sec3.word.Resize") #取HTML標中的 <div class="title"></div> 中的<a>標籤存入sel
    s = str(sel)
    s_replace = s.replace('<br>',"\n")
    print(s_replace) 
```
